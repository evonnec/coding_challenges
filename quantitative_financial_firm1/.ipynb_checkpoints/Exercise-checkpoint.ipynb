```python
import pandas as pd
import numpy as np
import collections
```

```python
trades_df = pd.read_csv("SIX_Trades.csv")
bluechips_df = pd.read_csv("SIXBlueChips.csv")
#reading in files
```

```python
trades_df_new = trades_df.merge(bluechips_df, on='RIC',how='outer')
#merge trades dataframe with blue chips dataframe. 
#on key is the RIC column
#use outer so there's no deletion of rows with NaN
```

```python
trades_df_new['Blue chip'] = trades_df_new['Blue chip'].fillna(False)
#fill in NA with False boolean for the new column Blue Chip
```

```python
trades_df_new['Blue chip'].value_counts()
#check that the boolean value for non Blue Chip now exists
```

```python
trades_df_new['mkt_value'] = trades_df_new['Shares'] * trades_df_new['Price (CHF)']
#make a new column mkt value
```

```python
comms = collections.OrderedDict([('mkt_value', 'sum')])
#takes the market value of each line and sums
```

```python
comms_per_order = trades_df_new.groupby(['OrderID', 'Blue chip']).agg(comms)
#groups by OrderID and whether it's a Blue Chip, aggregates the market value
```

```python
comms_per_order.info()
#take a look at what's in the new dataframe that lists mkt value per order
```

```python
#trades_df_new['vari_costs'] = [(0.0001 * trades_df_new['mkt_value']) if trades_df_new['Blue chip'] == True, else (trades_df_new['mkt_value'] * 0.0002)]

#trades_df_new['mkt_value'] * 0.0002 > 1

#trades_df_new['vari_costs'] = trades_df_new['Blue chip'].apply(lambda x: 0.0001 if x == True else 0)

#trades_df_new['vari_costs'] = trades_df_new['vari_costs'] * trades_df_new['mkt_value']

#these didn't work. need to define function as there are nested ifs in the function, see below
```

```python
def variable_costs(df,BC,netCHF):
    if df[BC] == True: return df[netCHF] * 0.0001
    if (df[netCHF] * 0.0002) < 1.0: return 1.0
    elif (df[netCHF] * 0.0002) > 1.0: return (df[netCHF] * 0.0002)
    else: return 0.0
#function of variable costs given a data frame, Blue Chip and mkt value
```

```python
comms_per_order = comms_per_order.reset_index()
#it has a two-pronged index. reset it.
```

```python
comms_per_order.info()
#take a look at the info on the dataframe
```

```python
comms_per_order['vari_costs'] = comms_per_order.apply(variable_costs, args=('Blue chip','mkt_value'),axis=1)
#apply the variable costs function, the name of the new column in comms_per_order df is 'vari costs',
#give it the args that the funciton requires, Blue Chip column and mkt_value column
```

```python
comms_per_order
#show the df with the new column and the variable costs per OrderID
```

```python
variable_costs = comms_per_order['vari_costs'].sum()
#sum the variable costs
```

```python
fixed_costs = len(comms_per_order) * 0.01
#determine fixed costs. take length and multiply by 0.01 CHF
```

```python
all_costs = "{:,}".format(fixed_costs + variable_costs)
#formatted costs - both variable and fixed
```

```python
all_costs = 'Total Fees Incurred: ' + str(all_costs)
#give it an introductory string
```

```python
num_orders = 'Total Number of Orders: ' + str(len(comms_per_order))
#number of orders with formatting
```

```python
shares_traded = 'Total Number of Shares Traded: ' + "{:,}".format(trades_df_new['Shares'].sum())
#shares traded with formatting
```

```python
total_notional_CHF_traded = 'Total Notional CHF Traded: ' + str("{:,}".format(trades_df_new['mkt_value'].sum()))
#notional CHF with formatting
```
